```python
for col in df.columns:
    plt.figure(figsize=(10, 4))
    if df[col].dtype == 'object':
        df[col].value_counts().plot(kind='bar', color='green')
        plt.title(f'Value Counts of {col}')
        plt.ylabel('Count')
    else:
        sns.histplot(df[col], kde=True, bins=30, color='blue')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Relationship between features and target
target = 'Efficiency_Status'
numerical_features = df.select_dtypes(include=np.number).columns
```

```python
for col in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=target, y=col, data=df)
    plt.title(f'{col} vs {target}')
    plt.tight_layout()
    plt.show()
```

```python
# Countplot for Operation_Mode vs Efficiency_Status
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Operation_Mode', hue='Efficiency_Status')
plt.title('Operation Mode vs Efficiency Status')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.countplot(data=df, x='Machine_ID', hue='Efficiency_Status')
plt.title('Operation Mode vs Efficiency Status')
plt.tight_layout()
plt.show()
```
